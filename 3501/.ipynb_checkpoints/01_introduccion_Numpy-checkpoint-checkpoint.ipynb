### Arrays

un **Array** es una estructura de datos que conciste en una coleccion de elementos (valores o variables) cada uno identificado por un indice o clave, un array se almacena de modo que la posicion de cada elemento se pueda calcular, a partir de su tupla de indice mediante su formula matematica. el tipo mas simple del array en un array lineal. tambien llamado unidimensional


En Numpy:

* cada dimencion se denomina **axis**.
* el numero de dimensiones se denomina **rank**.
* la lista de dimensiones con su correspondiente longitu se denomina **shape**.
* el numero total de elementos (multiplicacion de la longitud de las dimensiones) se denomina **size**.

```python
# Array con  cuyos valores son todos ceros 

a = np.zeros((2,4))
```

```python
a
```

**a** es un Array:
* con dos **axis** la primera con longitud 2 y el segundo con longitud 4
* con un **rank** igual a 2.
* con un **shape** igual (2,4)
* con un **size** igual a 8

```python
a.shape
```

```python
a.ndim
```

```python
a.size
```

### acceso a los elementos de un array 

#### Array Unidimensional

```python
# crear un arary unidemensional 

array_uni = np.array((1,3,5,7,9,11))

print("shape:", array_uni.shape)
print ("array uni: ", array_uni)
```

```python
# acceder al quinto elemento del array 

array_uni[4]
```

```python
# acceder al  tercer y cuato elemento del array 

array_uni[2: 4]
```

```python
# acceder al elemento 0,3,5 del array 

array_uni[[0,3,5]]
```

#### array multidimensional

```python
# crear un array multidimensional 

array_multi=np.array(([1,2,3,4], [5,6,7,8]))

print("shape: ", array_multi.shape)
print("Array_multi : \n", array_multi)
```

```python
# acceder al cuaro elemento del array 

array_multi[0, 3]
```

```python
# acceder a la segunda fila del array 

array_multi[1, :]
```

```python
# yo quiero acceder a las dos primeras filas del array 

array_multi[0: 2, 2]
```

#### modificar una Array 


```python
# crear un an array unidimesional e inicializarlo con el rango de elementos del 0 al 27 

array1=np.arange(28)
print("Shape: ", array1.shape)
print("Array1 : \n", array1)
```

```python
# cambiar las dimensiones del array y sus longitudes 

array1.shape=(7,4)
print("Shape: ", array1.shape)
print("Array1 : \n", array1)
```

```python
# el ejemplo anterior devuelve un nuevo aray que apunta a los mismos datos 
# nota, las modificaciones de un array modifican el otro array 

array2 = array1.reshape(4,7)
print("Shape: ", array2.shape)
print("Array2 : \n", array2)
print("Array1 : \n", array1)
```

```python
# modificar el array devuelto 

array2[0, 3] = 20
print("Array2 : \n", array2)
print("Array1 : \n", array1)
```

```python
# regresar el array devolviendo un nuevo array de una sola dimension
# nota el nuevo array apunta a los mismos datos 
print("array 1 : ", array1.ravel())
```

#### operaciones aritmeticas con arrays

```python
#creacion de ados arrays unidimensionales 

array1= np.arange(2, 18, 2)
array2= np.arange(8)
print ("array 1:", array1)
print("array 2:", array2)
```

```python
#suma
print (array1 + array2)
```

```python
#resta 

print (array1 - array2)
```

```python
# multiplicacion 

print (array1 * array2)
```

# Broadcating

#### si se aplican operaciones aritmeticas sobre arrays que no tengan la misma forma **shape** Numpy aplica una propiedad que se denomina Broadcasting.

```python
#creacion de dos arrays unidimensionales 

array1= np.arange(5)
array2= np.array([3])
print ("shape array 1:", array1.shape)
print("array 1 :", array1)
print("\n")
print ("shape array 2:", array2.shape)
print("array 2 :", array2)
```

```python
# suma de ambos aarregos 

array1 + array2 
```

```python
# crear un arreglo Multidimensional y unidimensional

array1= np.arange(6)
array1.shape = (2, 3)

array2= np.arange (6, 18, 4)
print ("shape array 1:", array1.shape)
print("array 1 :", array1)
print("")
print ("shape array 2:", array2.shape)
print("array 2 :", array2)
```

```python
# suma de ambos arrays 

array1 + array2
```
